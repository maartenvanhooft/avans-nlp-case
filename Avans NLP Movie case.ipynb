```python
import pandas as pd
import spacy
import numpy as np
pd.set_option("display.max_rows", 999)
pd.set_option("display.max_columns", 999)
pd.set_option("display.max_colwidth", 9999)
```

```python
df = pd.read_csv('movies.csv')
df["overview"] = df["overview"].astype(str)
corpus = df["overview"].to_list()
```

```python
df.head()
```

```python
from numpy import dot
from numpy.linalg import norm


def cosine_similarity(a,b):
    cos_sim = dot(a, b)/(norm(a)*norm(b))
    return cos_sim

def get_closest(df, ix):
    base_embed = df.at[ix, "embedding"]
    print(f"Find closest movie for {df.at[ix, 'title']}: \n\n {df.at[ix, 'overview']}")
    
    for i in df.index:
        this_embed = df.at[i, "embedding"]
        df.at[i, "similarity"] = cosine_similarity(base_embed, this_embed)
        
    return df.sort_values("similarity", ascending=False)[["title", "overview", "similarity"]]
```
